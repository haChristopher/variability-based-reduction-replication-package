```python
import matplotlib.pyplot as plt
```

```python
path = "/Users/christopher/Uni/MasterThesis/repositories/roaring/results.txt"
path2 = "/Users/christopher/Uni/MasterThesis/repositories/roaring/results1.txt"
```

```python
# read in file line by line
data = []
with open(path, "r") as f:
    lines = f.readlines()
    for line in lines:
        if "ns/op" in line:
            value = line.split("ns/op")[0].strip()
            value = value.split(" ")[-1]
            data.append(float(value))
```

```python
# read in file line by line
data2 = []
with open(path2, "r") as f:
    lines = f.readlines()
    for line in lines:
        if "ns/op" in line:
            value = line.split("ns/op")[0].strip()
            value = value.split(" ")[-1]
            data2.append(float(value))
```

```python

# Create a histogram
plt.hist(data, bins=200, edgecolor='black')
plt.xlabel('mean execution time per operation (ns)')
plt.ylabel('result frequency')
plt.title('BenchmarkShortIteratorNextBitmap/init_size_1_shortIterator_next-8')
plt.grid(axis='y', alpha=0.75)
plt.xlim(300, 1000)
plt.ylim(0, 150)

# Display the histogram
plt.show()
```

```python
 # Create subplots with 1 row and 2 columns
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Plot the first graph on the left subplot
ax1.hist(data, bins=200, edgecolor='black')
ax1.set_xlabel('mean execution time per operation (ns)')
ax1.set_ylabel('result frequency')
ax1.set_title('BenchmarkShortIteratorNextBitmap/init_size_1_shortIterator_next-8')
ax1.grid(axis='y', alpha=0.75)
ax1.set_xlim(300, 1000)
ax1.set_ylim(0, 150)

# Plot the second graph on the right subplot
ax2.hist(data2, bins=100, edgecolor='black')
ax2.set_xlabel('mean execution time per operation (ns)')
ax2.set_ylabel('result frequency')
ax2.set_title('BenchmarkChecksum/checksum-256')
ax2.grid(axis='y', alpha=0.75)

# Adjust the layout to prevent overlapping
plt.tight_layout()

# Display the plots
plt.savefig('benchmark_distributions.png', dpi=300)
```
