```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1)) 
```

```python
path = "/Users/christopher/Uni/MasterThesis/repositories/roaring/results.txt"
path2 = "/Users/christopher/Uni/MasterThesis/repositories/roaring/results1.txt"
```

```python
# read in file line by line
data = []
with open(path, "r") as f:
    lines = f.readlines()
    for line in lines:
        if "ns/op" in line:
            value = line.split("ns/op")[0].strip()
            value = value.split(" ")[-1]
            data.append(float(value))
```

```python
# read in file line by line
data2 = []
with open(path2, "r") as f:
    lines = f.readlines()
    for line in lines:
        if "ns/op" in line:
            value = line.split("ns/op")[0].strip()
            value = value.split(" ")[-1]
            data2.append(float(value))
```

```python

# Create a histogram
plt.hist(data, bins=200, edgecolor='black')
plt.xlabel('mean execution time per operation (ns)')
plt.ylabel('result frequency')
plt.title('BenchmarkShortIteratorNextBitmap/init_size_1_shortIterator_next-8')
plt.grid(axis='y', alpha=0.75)
plt.xlim(300, 1000)
plt.ylim(0, 150)

# Display the histogram
plt.show()
```
